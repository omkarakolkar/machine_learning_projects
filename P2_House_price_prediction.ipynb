# House price prediction

importing dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
import urllib.request
```

Importing data set

```python
#parsing data from url and converting into pandas format
url = "http://lib.stat.cmu.edu/datasets/boston"
raw_data = urllib.request.urlopen(url).read().decode('utf-8').split('\n')

# Extract only the lines containing actual data
data_lines = raw_data[22:]  # skip header lines

# Each sample has 13 features + 1 target value
# The data is structured with two rows per sample
data = []
for i in range(0, len(data_lines) - 1, 2):
    row1 = list(map(float, data_lines[i].strip().split()))
    row2 = list(map(float, data_lines[i+1].strip().split()))
    data.append(row1 + row2)
# data_lines[i] gets the i-th line (first line of the pair).
# .strip() removes any spaces or newlines at the beginning or end of the line.
# .split() splits the string into a list of strings wherever it finds spaces.
# map(float, ...) converts each string to a float number.
# list(...) wraps the result in a list.

# Convert to DataFrame
df = pd.DataFrame(data)

# Assign column names (based on the original dataset)
df.columns = [
    'CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE',
    'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT', 'Price'
]

# Preview
df.head()
```

# Splitting data sets

```python
X = df.drop(['Price'], axis=1)
Y = df['Price']
```

```python
#training and splitdata

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=2)
```

# Model training

```python
#loading the model
model = XGBRegressor() #decision trees with XGboost a type of tree ensemble

model.fit(X_train, Y_train)
```

# Evaluation

```python
# accuracy for prediction

training_data_prediction = model.predict(X_train)

print(training_data_prediction)
```

```python
# R squared error for regression and for classification is accuracy score
# we will be use using R squared
score_1 = metrics.r2_score(Y_train, training_data_prediction)
# The value of R-square lies between 0 to 1. Where we get R-square equals 1 when the model perfectly fits the data and there is no difference
# between the predicted value and actual value.  However, we get R-square equals 0 when the model does not predict any variability in the model
# and it does not learn any relationship between the dependent and independent variables.



#mean absolute error If it's <10% of the mean of your targets, that's often considered reasonable
score_2 = metrics.mean_absolute_error(Y_train, training_data_prediction)

print("RsE: ",score_1)
print("MAE: ",score_2)
```

# Visualising actual and prediction

```python
plt.scatter(Y_train, training_data_prediction)
plt.xlabel("actual")
plt.ylabel("predict")
plt.show()
```

# Prediction on test data

```python
test_data_prediction = model.predict(X_test)
```

```python
#r squared error
testscore_1 = metrics.r2_score(Y_test, test_data_prediction)
# mean absolute error
testscore_2 = metrics.mean_absolute_error(Y_test, test_data_prediction)

print("RsE: ",testscore_1)
print("MAE: ",testscore_2)
```
