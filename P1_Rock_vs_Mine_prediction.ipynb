```python
file_path = '/content/drive/MyDrive/Machine Learning project /datasets/sonar data.csv'

import pandas as pd

df = pd.read_csv(file_path)
df.columns = range(df.shape[1]) # to add column headers from  0-60
df.head()
```

```python
import numpy as np #importing libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# Splitting data

```python
#seperating data and targets
X = df.drop(columns = 60, axis = 1)
Y = df[60]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state = 1)
#stratify=Y it will split about equal number of M & R into training and test data
#random state variable
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# Model Training Logistic Regression model

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
#acc on training data
X_train_prediction = model.predict(X_train)
training_data_acc = accuracy_score(X_train_prediction, Y_train)

print(training_data_acc)
```

```python
#acc on test data
X_test_prediction = model.predict(X_test)
test_data_acc = accuracy_score(X_test_prediction, Y_test)

print(test_data_acc)
```
